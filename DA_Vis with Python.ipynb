#                                Data Visualisation Using Python

```python
import pandas as pd # Pandas will be used to import data from CSV files
from matplotlib import pyplot as plt # Import pyplot from the matplotlib module (Used to make charts)
```

```python
# Example of pyplot

# Define inputs
x = [1,2,3]
y = [1,4,9]
z = [10,5,0]


# Plot the graph
plt.plot(x,y) # Plot 1 of 2 lines
plt.plot(x,z) # Plot 2 of 2 lines
plt.title("Test Plot") # Add title
plt.xlabel("X") # Add xlabel
plt.ylabel("Y and Z") # Add ylabel
plt.legend(["This is Y", "This is Z"]) # Add legend to define each line
plt.show() # Show plot
```

```python
data = pd.read_csv("countries.csv") # Load the counntries CSV file and store in Variable called data
```

```python
print(data) # View data in DataFrame
```

## Compare the population growth in the US and China

```python
us = data[data.country == "United States"] # Assign all US entries to a variable called US
```

```python
china = data[data.country == "China"] # Assign all China entries to a variable called China
```

```python
plt.plot(us.year, us.population / 10 **6) # Display growth of US population by year(10**6 will display the pop in millions)
plt.plot(china.year, china.population / 10**6)
plt.legend(["United States", "China"]) # Add legend to define each line
plt.xlabel("Year") # Add x label
plt.ylabel("Population") # Add ylabel
plt.show() # Show Graph
```

### Show in percentage amounts

```python
# To achieve this we need to divided the population by the 1st number and multiple by 100
percentageUS = us.population / us.population.iloc[0] * 100

percentageChina = china.population / china.population.iloc[0] * 100
```

```python
plt.plot(us.year, percentageUS) # Display growth of US population by year(10**6 will display the pop in millions)
plt.plot(china.year, percentageChina)
plt.legend(["United States", "China"]) # Add legend to define each line
plt.xlabel("Year") # Add x label
plt.ylabel("Population Growth (First Year = 100 %)") # Add ylabel
plt.show() # Show Graph
```

#### From the graph we can see that Chinas population growth was much faster than the US. The US population growth was 90% while Chinas growth was over 120%
